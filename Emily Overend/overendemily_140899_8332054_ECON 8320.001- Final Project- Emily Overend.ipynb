```python
# This is the result of my first attempt at scraping the data from the NIL College Athletes website.
# The function outputs a data frame, but it does not scrape data from the other pages and put it in the data frame.
# It also separates each sponsor, so the first 12 sponsors actually belong to the first student listed.
# I'm not sure why it does not scrape data from the rest of the pages, and I'm also not sure why it separates each sponsor. I have a separate loop that separates each individual sponsor shown below.

import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
from urllib.parse import urljoin

nilurl= "https://nilcollegeathletes.com/athletes"

def scrapeNIL(nilurl):
    
    #Retrieving the NIL website:
    nilcolathel = requests.get("https://nilcollegeathletes.com/athletes")
    
    #Parsing the website with BeautifulSoup:
    parsed = BeautifulSoup(nilcolathel.text)
    
    #Names:
    nameParsed = [i for i in parsed.find_all('td', class_="px-2 md:px-6 py-4 whitespace-nowrap text-sm font-medium text-gray-900")]
    
    #Sponsors:
    sponsorParsed= parsed.find_all('li', class_="text-sm text-gray-900 truncate")
    
    #University and Sport
    b = [i for i in parsed.find_all('span', class_="truncate")]

    #Iterating over all names:
    studNames=[]
    for i in nameParsed:
        #Add the name to the names list:
        studNames.append(i.text.strip())
    studNames=list(studNames)
    studNames= pd.DataFrame(studNames, columns=['Name'])
    #studNames
    
    #Sponsors:     
    sponsors= [i.text.strip().replace('\n', '') for i in sponsorParsed]
    sponsors=list(sponsors)
    sponsors=pd.DataFrame(sponsors,columns=['Sponsor'])
    #sponsors
    
    #University and Sport
    uni= [i.text.strip() for i in b]
    #uni
    
    univ=[]
    sport=[]
    
    # Separating university and sport:
    for i, j in enumerate(uni): #enumerate adds a counter to an iterable and returns it to a form of enumerating object
        if i%2==0: #If the remainder of the element is 0 then it appends the value to the univ list.
            univ.append(j.strip())
        else:
            sport.append(j.strip()) #If the remainder of the element is not 0, then it appends it to the sport list. 
    
    #Creating a data frame with university and sport:
    univ=list(univ)
    sport=list(sport)
    univ= pd.DataFrame(univ, columns=['university'])
    sport= pd.DataFrame(sport, columns=['sport'])
    univSport= univ.join([sport])
    #univSport
    
    #Creating a dataframe
    nilDf= studNames.join([sponsors, univSport])
    
    return nilDf
    
    #Getting data from other pages:
    root='https://nilcollegeathletes.com'
    try:
        nextPage= urljoin(root, parsed.find('div', class_='-mt-px flex w-0 flex-1 justify-end').a['href']) #This joins the root object and the href
    except:
        nextPage= None
    
    #Combining data from other pages (If there is another page of results):
    if nextPage:
        return pd.concat([nilDf, scrapeNIL(nextPage)], axis=0)
    #Otherwise return the current data
    else:
        return nilDf
```

```python
# Scrapping the On3 NIL 100

nil100url='https://www.on3.com/nil/rankings/player/nil-100/'

# Function that changes the social media and valuation values to floats
def valchange(i):
    if i[-1:] ==  'K':  # Check if the last digit is K
        return float(i[:-1])*1000 # Remove the last digit with [:-1], and convert to int and multiply by 1000
    elif i[-1:] == 'M':  # Check if the last digit is M
        return float(i[:-1])* 1000000 # Remove the last digit with [:-1], and convert to int and multiply by 1000000
    else:  # just in case data doesnt have an M or K
        i= float(i)  # just in case data doesnt have an M or K
        
def scrapeNIL100(nil100url):
    # Retrieving the On3 NIL 100 Page
    nil100= requests.get('https://www.on3.com/nil/rankings/player/nil-100/')

    #Parsing the page with Beautiful Soup
    nil100Parsed= BeautifulSoup(nil100.text)

    # School Level (College or High School)
    schoolLevelparsed= [i for i in nil100Parsed.find_all('h6', class_='MuiTypography-root NilPlayerRankingItem_orgType__jTK2b MuiTypography-subtitle1 MuiTypography-colorTextPrimary')]

    # Athlete's Name
    athleteNameparsed= [i for i in nil100Parsed.find_all('div', class_='NilPlayerRankingItem_nameYearContainer__R16k_')]

    #Rating
    Ratingsparsed= [i for i in nil100Parsed.find_all('span', class_='StarRating_overallRating__MTh52 StarRating_bolded__kr_6V StarRating_border__DffWl')]

    #Status
    statusParsed= [i for i in nil100Parsed.find_all('p', class_='MuiTypography-root NilPlayerRankingItem_statusText__G8SdI MuiTypography-body1 MuiTypography-colorTextPrimary')]
    
    #Social Media Followers
    socMediaparsed= [i for i in nil100Parsed.find_all('p', class_='MuiTypography-root NilPlayerRankingItem_followersNumber__ifWQr MuiTypography-body1 MuiTypography-colorTextPrimary')]
    
    #On3 NIL Valuation
    valParsed= [i for i in nil100Parsed.find_all('p', class_='MuiTypography-root NilPlayerRankingItem_valuationCurrency__oSkvo MuiTypography-body1 MuiTypography-colorTextPrimary')]
    
    # Iterating over the School Level 
    schoolLevel=[]
    for i in schoolLevelparsed:
        schoolLevel.append(i.text.strip())
    schoolLevel= pd.DataFrame(schoolLevel, columns=['school_level'])
    #schoolLevel

    # Iterating over the Athlete's Name
    athleteName=[]
    for i in athleteNameparsed:
        athleteName.append(i.text.strip())
    athleteName= pd.DataFrame(athleteName, columns=['athelete_name'])
    #athleteName
    
    # Combining schoolLevel and athleteName into one data frame:
    col12= schoolLevel.join([athleteName])

    #Iterating over the Ratings
    ratings=[]
    for i in Ratingsparsed:
        ratings.append(i.text.strip())
    ratings= pd.DataFrame(ratings, columns=['ratings'])
    #Ratings
    
    # Adding Ratings to the data frame:
    col123= col12.join([ratings])
    
    # Iterating over the Status 
    status=[]
    for i in statusParsed:
        status.append(i.text.strip())
    status= pd.DataFrame(status, columns=['status'])
    #status
    
    # Adding Status to the data frame:
    col1234= col123.join([status])
    
    # Iterating over Social Media followers
    socialMedia=[]
    for i in socMediaparsed:
        i_soctext= i.text.strip()
        i_socfloat= valchange(i_soctext)
        socialMedia.append(i_socfloat)
    socialMedia= pd.DataFrame(socialMedia, columns=['soc_med_followers'])
    #socialMedia
    
    # Adding socialMedia to the data frame:
    col12345= col1234.join([socialMedia])
    
    # Iterating over On3 NIL Valuation
    valuation=[]
    for i in valParsed:
        i_text= i.text.strip()
        i_nodolla= i_text.replace('$', '')
        i_valfloat= valchange(i_nodolla)
        valuation.append(i_valfloat)
    valuation= pd.DataFrame(valuation, columns=['valuation'])
    #valuation
    
    # Adding valuation to the data frame (final data frame):
    nil100Df= col12345.join([valuation])
    
    return nil100Df

on3nil100= scrapeNIL100('https://www.on3.com/nil/rankings/player/nil-100/')
on3nil100
```

```python
# Multiple Regression of the On3 NIL 100 data with Valuation as the dependent variable:
import statsmodels.formula.api as smf

reg = smf.ols("valuation ~ C(school_level) + C(status) + soc_med_followers", data=on3nil100).fit()
print(reg.summary())
```
