```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import html5lib
```

```python
# Create empty list of scraped deals
list_deals = []

# For the pages in the range, go to the URL in successive pages using the user agent to appear as a browser to the site
for page_number in range(1, 421):
    urlopen_req = requests.get(f"https://nilcollegeathletes.com/athletes?page={page_number}", 
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}).text

    # Append each successive page's info
    list_deals.append(pd.read_html(urlopen_req, flavor="lxml")[0].assign(PageNumber= page_number))

# Concatenate the results of making each row of data have info for each sponsor, even if for the same athlete
nildeals = (pd.concat(list_deals, ignore_index=True)
    .assign(Sponsors= lambda x: x["Sponsors"].str.split(r"\s\s+"))
    .explode("Sponsors"))

nildeals
```

```python
nildeals.to_csv(r'Desktop\nildeals.csv', index=False)
```
